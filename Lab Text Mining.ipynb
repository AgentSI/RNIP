```python
# Part I. Text processing
# 1. Tokenisation
import nltk
nltk.download('punkt_tab')

text = "Your professor Nistor Grozavu is a member of the ETIS laboratory. The ETIS laboratory consists of 5 teams and has 150 researchers. Nistor's main research work is data mining and machine learning. Machine Learning is also used in Text Mining and word research."

token = nltk.word_tokenize(text)
token
```

```python
# 2. Finding frequency distinct in the text
from nltk.probability import FreqDist

fdist = FreqDist(token)
k = 8
fdist1 = fdist.most_common(k)
fdist1
```

```python
# 3. Stemming
from nltk.stem import PorterStemmer
pst = PorterStemmer()
pst.stem("research")
stm = ["research", "researching", "researchers"]
for word in stm :
   print(word+ ":" +pst.stem(word))

# Test for 2 other words.
pst.stem("house")
stm = ["house", "houses", "housing"]
for word in stm :
   print(word+ ":" +pst.stem(word))

pst.stem("friend")
stm = ["friend", "friendship", "friends", "friendships"]
for word in stm :
   print(word+ ":" +pst.stem(word))
```

```python
# 4. Lemmatisation
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print("rocks :", lemmatizer.lemmatize("rocks"))
print("corpora :", lemmatizer.lemmatize("corpora"))

# Test for other words.
print("is :", lemmatizer.lemmatize("is"))
print("only :", lemmatizer.lemmatize("only"))
print("realized :", lemmatizer.lemmatize("realized"))
```

```python
# 5. Stop Words
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words("english"))
text = "Cristiano Ronaldo was born on February 5, 1985, in Funchal, Madeira, Portugal."
text1 = nltk.word_tokenize(text.lower())
print(text1)
stopwords = [x for x in text1 if x not in stop_words]
print(stopwords)
```

```python
# Part II. Clustering text data
# 1. Import wikipedia data.

import pandas as pd
import wikipediaapi

wiki_wiki = wikipediaapi.Wikipedia(
    language='en',
    user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
)

articles=['Data Science','Artificial intelligence','Machine Learning','European Central Bank','Bank','Financial technology','International Monetary Fund','Basketball','Swimming','Tennis']

wiki_lst=[]
title=[]

for article in articles:
   print("loading content: ", article)
   page = wiki_wiki.page(article)
   wiki_lst.append(page.text)
   title.append(article)

print("examine content")
wiki_lst
```

```python
# 2. Represent each article as a vector
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load 4 categories for example:
categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']

# Fetch the dataset from sklearn
from sklearn.datasets import fetch_20newsgroups
twenty_data = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42)

# Represent each article as a vector (using TF-IDF vectorization)
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(wiki_lst)

# Perform k-means clustering
Sum_of_squared_distances = []
K = range(2,10)

for k in K:
   km = KMeans(n_clusters=k, max_iter=200, n_init=10)
   km = km.fit(X)
   Sum_of_squared_distances.append(km.inertia_)

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# We will try to cluster into 6 groups.
true_k = 6
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=10)
model.fit(X)
labels=model.labels_
wiki_cl=pd.DataFrame(list(zip(title,labels)),columns=['title','cluster'])
print(wiki_cl.sort_values(by=['cluster']))
```

```python
# 4. Evaluate the result
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

result={'cluster':labels,'wiki':wiki_lst}
result=pd.DataFrame(result)

for k in range(0, true_k):
   # Get the rows for the current cluster
   s=result[result.cluster == k]

   # Concatenate all the text for this cluster
   text=s['wiki'].str.cat(sep=' ')
   text=text.lower()
   text=' '.join([word for word in text.split()])

   # Generate the word cloud
   wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)

   # Print the cluster label and the corresponding titles of the articles
   print('Cluster: {}'.format(k))
   print('Titles')
   titles=wiki_cl[wiki_cl.cluster==k]['title']
   print(titles.to_string(index=False))

   # Display the word cloud
   plt.figure()
   plt.imshow(wordcloud, interpolation="bilinear")
   plt.axis("off")
   plt.show()
```
